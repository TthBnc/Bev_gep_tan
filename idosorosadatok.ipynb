```python
def build_model():
    inputs = keras.Input(shape=(2,2))
    x = LSTM(256,return_sequences=True)(inputs)
    x = LSTM(256,return_sequences=True)(x)
    x = Dropout(0.2)(x)
    x = LSTM(256,return_sequences=True)(x)
    x = Dropout(0.2)(x)
#x = LSTM(256,return_sequences=True)(x)
    x = GlobalAveragePooling1D()(x)
    x = Dense(64,activation="relu")(x)
    x = Dropout(0.2)(x)
    x = Dense(32,activation="relu")(x)
    x = Dropout(0.2)(x)
    out = Dense(1,name='out1',activation="linear")(x)
    model = Model(inputs=inputs, outputs=out)
    return model
    model.compile(optimizer='adam', loss={
	    "out1":'rmse',
        #"out2":'sparse_categorical_crossentropy'#tf.keras.losses.SparseCategoricalCrossentropy()#"sparse_categorial_crossentropy"#tfk.losses.CategoricalCrossentropy()#(from_logits=True)#segmentation_loss
    }, metrics=['acc'])
```

```python
model = build_model()
```

```python
adatok = []
y = []
for i in osszadatok:
    for j in range(0,len(i)-2):
        #print(i[j:j+2])
        adatok.append(i[j:j+2])
        y.append(i[j+2][0])
```

```python
model.fit(x=adatok,y=y)
```
